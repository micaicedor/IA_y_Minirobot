```python
# =====================================
# 游댳 2. IMPORTACIONES
# =====================================
from langchain_community.llms import Ollama
from langchain_ollama import OllamaEmbeddings
from langchain_chroma import Chroma
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate
from langchain.chains import create_retrieval_chain
import requests
from bs4 import BeautifulSoup
import os
```

```python

# =====================================
# 游댳 3. CONFIGURACI칍N DEL MODELO
# =====================================
llm_llama = Ollama(model="llama3.1")
embeddings = OllamaEmbeddings(model="nomic-embed-text")
persist_directory = "chroma_unal"
```

```python
# =====================================
# 游댳 4. FUNCI칍N PARA EXTRAER NOTICIAS
# =====================================
def get_unal_news(limit=10):
    """
    Extrae las noticias m치s recientes de la Agencia de Noticias UNAL.
    Devuelve una lista de diccionarios con t칤tulo, resumen y enlace.
    """
    url = "https://agenciadenoticias.unal.edu.co/"
    news_items = []
    try:
        resp = requests.get(url, timeout=10)
        resp.raise_for_status()  # Raise an exception for bad status codes
        soup = BeautifulSoup(resp.text, "html.parser")

        # Updated selector based on user's input and common news article structure
        articles = soup.select("article.article")
        if not articles:
            print(f"Advertencia: No se encontraron elementos con el selector 'article.article' en {url}")
            # Fallback to a broader selector if the specific one fails
            articles = soup.select("article")
            if not articles:
                 print(f"Advertencia: No se encontraron elementos con el selector 'article' en {url}")
                 return []


        for art in articles[:limit]:
            # Attempt to find title and summary within the article element
            title_tag = art.select_one("h3 a, h2 a, h5 a") # Look for common title tags within article
            summary_tag = art.select_one("p.lead, p.card-text, .lead") # Look for common summary/lead paragraph tags

            if not title_tag:
                print("Advertencia: Se encontr칩 una tarjeta de noticia sin t칤tulo.")
                continue

            title = title_tag.get_text(strip=True)
            # Construct the link carefully, handling relative and absolute paths
            link = title_tag.get("href")
            if link and not link.startswith("http"):
                 link = "https://agenciadenoticias.unal.edu.co" + link


            summary = summary_tag.get_text(strip=True) if summary_tag else ""

            content = f"T칤tulo: {title}\nResumen: {summary}\nEnlace: {link}"
            news_items.append(content)
            print(f"Noticia extra칤da: {title}") # Debug print

    except requests.exceptions.RequestException as e:
        print(f"Error al intentar acceder a la p치gina: {e}")
        return []
    except Exception as e:
        print(f"Ocurri칩 un error durante el scraping: {e}")
        return []

    return news_items
```

```python
# =====================================
# 游댳 5. CREAR O CARGAR VECTOR STORE
# =====================================
vector_store_exists = os.path.exists(persist_directory)
vector_store = None
if vector_store_exists:
    print("Cargando base de datos local existente...")
    try:
        vector_store = Chroma(
            collection_name="unal_news",
            embedding_function=embeddings,
            persist_directory=persist_directory
        )
        if len(vector_store.get()["ids"]) > 0:
            print("Documentos almacenados en la base local:", len(vector_store.get()["ids"]))
        else:
            print("La base de datos local est치 vac칤a. Procediendo a descargar noticias...")
            vector_store = None # Reset vector_store to trigger news download

    except Exception as e:
        print(f"Error al cargar la base de datos local: {e}")
        vector_store = None # Reset vector_store to trigger news download


if not vector_store:
    print("Descargando noticias recientes de la UNAL...")
    news_texts = get_unal_news(limit=15)
    if news_texts:
        for n, noticia in enumerate(news_texts, 1):
          print(f"游닗 Noticia {n}:\n{noticia}\n{'-'*60}")

        vector_store = Chroma.from_texts(
            texts=news_texts,
            collection_name="unal_news",
            embedding=embeddings,
            persist_directory=persist_directory
        )
        print(f"{len(news_texts)} noticias almacenadas.")
    else:
        print("No se pudieron descargar noticias de la UNAL.")
```

```python
# =====================================
# 游댳 6. CREAR RETRIEVER
# =====================================
retriever = vector_store.as_retriever(search_kwargs={"k": 3})
```

```python

# =====================================
# 游댳 7. DEFINIR PROMPT Y CADENA RAG
# =====================================
system_prompt = (
    "Eres un asistente especializado en la Universidad Nacional de Colombia. "
    "Responde usando informaci칩n verificada de la Agencia de Noticias UNAL. "
    "Si no tienes informaci칩n suficiente, responde con: 'No tengo informaci칩n actualizada sobre ese tema.'\n\n"
    "Contexto relevante:\n{context}"
)

prompt = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    ("human", "{input}"),
])

qa_chain = create_stuff_documents_chain(llm_llama, prompt)
rag_chain = create_retrieval_chain(retriever, qa_chain)
```

```python


# =====================================
# 游댳 8. CHAT INTERACTIVO
# =====================================
def chat_unal(query):
    response = rag_chain.invoke({"input": query})
    return response["answer"]
```
